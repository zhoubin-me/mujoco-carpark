```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_sb3 = pd.read_csv('~/Downloads/PPO_1 (1).csv')
```

```python
data_ours1 = pd.read_csv('~/Downloads/ppo_CarParkingMujocoPPO-v0_2025-04-26_01-20-24 (1).csv')
data_ours2 = pd.read_csv('~/Downloads/ppo_CarParkingMujocoPPO-v0_2025-04-26_02-29-42 (1).csv')
```

```python
plt.figure(figsize=(12, 9))
plt.plot(data_sb3['Step'], data_sb3['Value'].rolling(window=10).mean())
plt.plot(data_ours1['Step'], data_ours1['Value'].rolling(window=10).mean())
plt.plot(data_ours2['Step'], data_ours2['Value'].rolling(window=10).mean())
plt.xlabel('Env Step')
plt.ylabel('Episodic Reward')
plt.legend(['Stable Baselines3 PPO', 'Ours PPO run 1', 'Ours PPO run 2'])
plt.title('PPO Episodic Reward Comparison')
plt.savefig('PPO_Episodic_Reward.png')
```
